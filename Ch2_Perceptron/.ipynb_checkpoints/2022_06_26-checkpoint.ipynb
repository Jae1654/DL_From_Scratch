# 2.3 Implement Perceptron 

```python
def AND(x1, x2):
    w1, w2, theta = 0.5, 0.5, 0.7
    tmp = x1*w1 + x2*w2
    if tmp <= theta:
        return 0
    else:
        return 1

print(AND(0,0), AND(0,1), AND(1,0), AND(1,1))
```

###  2.3.2 Weight and Bias
Suppose we replace **ğœƒ** to **-b**, call **bias**

$$y=\begin{equation}
\left\{ 
  \begin{aligned}
    0(b + x_1*w_1 + x_2*w_2)\le0 \\
    1(b + x_1*w_1 + x_2*w_2)>0
  \end{aligned}
  \right.
\end{equation}
$$

Then, the perceptron produces 1 if the signals multiplied by wieghts and add bias b is greater than zero else 0

**bias** = The threshold for the activation of the neuron  

**weight** = The importance(effectiveness) of the signal to the output

```python
import numpy as np
x = np.array([0, 1]) # ì…ë ¥
w = np.array([0.5,0.5]) # ê°€ì¤‘ì¹˜
b = -0.7

np.sum(w*x) + b
```

### 2.3.3 Implement Weight and Bias

We can parametrize the perceptron

The **AND** circuit has three parameters (ğ‘¤1 , ğ‘¤2, b)

The **NAND** is the reverse of the **AND** (-ğ‘¤1 , -ğ‘¤2, -b)

The **OR** circuit is the modification of the **AND** (ğ‘¤1_* , ğ‘¤2_*, b_*)


```python
# parameters (0.5, 0.5, -0.7)
def AND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5])
    b = -0.7
    tmp = np.sum(w*x) + b
    
    if tmp <=0:
        return 0
    else:
        return 1
    
# parameters (-0.5, -0.5, 0.7)
def NAND(x1, x2):
    x = np.array([x1, x2])
    w = np.array([-0.5, -0.5]) # The reverse of the AND parameters
    b = 0.7
    tmp = np.sum(w*x) + b
    
    if tmp <=0:
        return 0
    else:
        return 1
    
# parameters (-0.5, -0.5, 0.7)
def OR(x1, x2):
    x = np.array([x1, x2])
    w = np.array([0.5, 0.5]) # The modification of the AND parameter
    b = -.2
    tmp = np.sum(w*x) + b
    
    if tmp <=0:
        return 0
    else:
        return 1
```

# 2.4 Limitation of the Perceptron

  \right.

$$
